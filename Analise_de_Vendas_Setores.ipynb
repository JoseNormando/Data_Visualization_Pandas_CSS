# Missão 1: Time de vendas

Precisamos construir uma visualização para indicar os **10 maiores clientes por total de venda**.

Com isso, a equipe de vendas pode pensar em estratégias para manter esses clientes comprando em nossa loja.

```python
df_cliente_vendas = df.groupby('nome_cliente')[['vendas']].sum().nlargest(10, columns = 'vendas').copy()
df_cliente_vendas = df_cliente_vendas.reset_index()
df_cliente_vendas
```

```python
df_cliente_vendas.columns = ['Clientes', 'Vendas']
df_cliente_vendas['Ranque'] = df_cliente_vendas.index + 1
df_cliente_vendas.set_index('Ranque', inplace = True)
df_cliente_vendas
```

```python
# Objeto Styler

s = df_cliente_vendas.style
s
```

```python
s.format({'Vendas': 'R$ {:,.2f}'})
s
```

# Missão 2: Time comercial

Precisamos construir uma visualização que permita associar o **total ganho em vendas e o total lucrado por tipo de produto**.

Isso vai permitir que o time comercial possa focar em estratégias para alavancar as rendas que ganham com cada tipo de produto vendido na loja.

```python
df_venda_lucro = df.groupby('tipo_produto')[['vendas', 'lucro']].sum()
df_venda_lucro.index.name = 'tipo produto'
df_venda_lucro
```

```python
estilo_produto = df_venda_lucro.style
estilo_produto
```

```python
# Já que estamos tratando agora apenas com colunas numéricas, podemos digitar direto a formatação

estilo_produto.format('R$ {:,.2f}')\
              .highlight_max(color = 'lightgreen')\
              .highlight_min(color = '#F16165')
estilo_produto
```

```python
estilo_produto.format('R$ {:,.2f}').background_gradient(cmap = 'Greens')
estilo_produto
```

```python
cabecalho = {
    'selector': 'th',  # seleciona o cabeçalho (titulo das colunas e index)
    'props': 'font-weight: bold; font-family: Arial; text-align: center; text-transform: capitalize;'
}
estilo_produto.set_table_styles([cabecalho], overwrite = False)
```

# Missão 3: Time de logística

Precisamos construir uma visualização que permita entender a **distribuição de pedidos por região do Brasil**.

Isso vai permitir que o time logística entenda onde é mais interessante focar material e pessoas.

```python
df_regiao = pd.DataFrame(df['regiao'].value_counts())
df_regiao.columns = ['N° pedidos']
df_regiao.index.name = 'Região'
df_regiao
```

```python
df_regiao['Porcentagem %'] = df['regiao'].value_counts(normalize = True) * 100
df_regiao
```

```python
estilo_regiao = df_regiao.style
estilo_regiao
```

```python
cabecalho_2 = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; text-align: right; background-color: black; color: #dddd55'
}
celulas = {
    'selector': 'td', # Se refere ao conteúdo da tabela
    'props': 'background-color: black;'
}

estilo_regiao.set_table_styles([cabecalho_2, celulas])

# A visualização white fica melhor no tema light
```

```python
estilo_regiao.format({'Porcentagem %': '{:.2f} %'})\
             .bar(subset = 'Porcentagem %', vmin = 0, vmax = 100.0, color = '#9CD33B')
estilo_regiao
```

# Missão 4: Time de logística

Novamente o time de logística deseja repensar sua estrutura e organização de fornecimento de produtos e para isso, foi pensada em uma visualização que permita identificar padrão de **quantidade de produtos solicitados por mês por departamento**.

```python
df_meses = df.copy()
df_meses = df_meses.sort_values('data_pedido')
df_meses
```

```python
df_meses['data_pedido'] = pd.to_datetime(df_meses['data_pedido'])
```

```python
df_meses['meses'] = df_meses['data_pedido'].dt.strftime('%Y - %b')
df_meses = df_meses.reset_index(drop = True)
df_meses.head()
```

```python
vendas_mensais = df_meses.pivot_table(index = 'departamento', columns = 'meses', values = 'quantidade',
                                      aggfunc = 'sum', sort = False)
vendas_mensais
```

```python
estilo_mensal = vendas_mensais.style
estilo_mensal
```

```python
estilo_mensal.set_sticky(axis = 'index')
```

```python
colunas = {
    'selector': '.col_headin', # Seleciona apenas o cabeçalho das colunas
    'props': 'font-weight: normal; font-family: Arial'
}
tabelas = {
    'selector': 'td,th', # Seleciona a tabela toda
    'props': 'text-align: left'

}
index = {
    'selector': '.index_name', # Seleciona apenas o nome do index
    'props': 'font-weight: bold; font-family: Arial; text-align: right;'
}
estilo_mensal.set_table_styles([colunas, tabelas, index], overwrite = False)
```

# Missão 5: Relatório de performance

O relatório de performance é um documento muito importante para a empresa. Nele, todos que trabalham na loja podem entender como anda a empresa. Por conta disso, tabelas bem construídas são essenciais para mostrar os dados da melhor forma.

Portanto, precisamos construir uma visualização que permita entender a **relação entre tipos de clientes e modo de envio de produtos de acordo com as vendas**. Com isso, será possível destacar no relatório qual o modo de envio e quais tipos de modo de envio mais geraram ganho em vendas.

```python
df_cliente = df.pivot_table(index = 'segmento_cliente', columns = 'modo_envio', values = 'vendas',
                            aggfunc = 'sum')
df_cliente
```

```python
# Criando coluna
df_cliente['Total'] = df_cliente.sum(axis = 1)

# Criando linha
df_cliente.loc['Total'] = df_cliente.sum()

df_cliente
```

```python
compra_cliente = df_cliente.style.format('{:,.2f}')
compra_cliente
```

```python
tabela = {
    'selector': 'td,th:not(.index_name)', # Não será considerado na formatação o nome dos index (segmento_cliente e modo_envio)
    'props': 'font-weight: normal; font-family: Arial; text-align: center; background-color: black'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: normal; text-align: right; font-style: italic; color: #696969'
}
compra_cliente.set_table_styles([tabela, index])
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': 'th',
        'props': 'border-top: 1px solid white'
    },
    {
     'selector': 'td',
     'props': 'border-top: 1px solid white'
    }],
    'B2B': [{
        'selector': 'th',
        'props': 'border-top: 1px solid white'
    },
    {
     'selector': 'td',
     'props': 'border-top: 1px solid white'
    }]
}, overwrite = False, axis = 1)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: purple'
    }]
}, overwrite = False, axis = 0)

cores_coluna = pd.DataFrame(['false', 'true', 'false'], index = df_cliente['Total'].index,
                            columns = ['Total'])

compra_cliente.set_td_classes(cores_coluna)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #006400'
    }]
}, overwrite = False, axis = 1)

cores_coluna_2 = pd.DataFrame([['false', 'false', 'true', 'false', 'false']], index = ['Total'], columns = df_cliente.columns)

compra_cliente.set_td_classes(cores_coluna_2)
```

```python
dados_datas = dados_loja.copy()
dados_datas.head()
```

```python
dados_datas['data_pedido'] = pd.to_datetime(dados_datas['data_pedido'])
dados_datas['data_chegada'] = pd.to_datetime(dados_datas['data_chegada'])

dados_datas = dados_datas.sort_values('data_pedido')
dados_datas = dados_datas.reset_index(drop = True)

dados_datas.head()
```

```python
dados_datas = dados_datas[dados_datas['data_pedido'] >= '2013-01-01']
dados_datas.head()
```

```python
dados_datas['meses'] = dados_datas['data_pedido'].dt.strftime('%Y - %b')
dados_datas.head()
```

```python
tempo_demora = (dados_datas['data_chegada'] - dados_datas['data_pedido']).dt.days
```

```python
dados_datas['tempo_entrega'] = tempo_demora
dados_datas.head()
```

```python
entregas = dados_datas.pivot_table(index = 'pais', columns = 'meses', values = 'tempo_entrega', aggfunc = 'mean', sort = False)
entregas
```

```python
s_entregas = entregas.style.set_sticky(axis = 0)
s_entregas
```
